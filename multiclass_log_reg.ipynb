```python
import numpy as np
import matplotlib.pyplot as plt

# vector inputs
phi1 = np.pi/4
length1 = 1
phi2 = 3*np.pi/4
length2 = 1


w1 = np.zeros([1,2])
w1[0,0] = np.cos(phi1)*length1
w1[0,1] = np.sin(phi1)*length1

w2 = np.zeros([1,2])
w2[0,0] = np.cos(phi2)*length2
w2[0,1] = np.sin(phi2)*length2

size = 121

steps = np.linspace(-2,2,num=size)

x = np.zeros([2,size*size])

for i in range(size):
    for j in range(size):
        x[0,size*i+j]=steps[i]
        x[1,size*i+j]=steps[j]
        
ew1 = np.exp(np.dot(w1,x))
ew2 = np.exp(np.dot(w2,x))

hw1 = ew1 / (ew1+ew2)
hw1 = hw1[0].reshape(size,size)
hw2 = ew2 / (ew1+ew2)
hw2 = hw2[0].reshape(size,size)

plt.rcParams.update({'font.size': 15})
plt.imshow(hw1.T,origin='lower',cmap="hot",vmin=0,vmax=1,extent = [-2,2,-2,2],aspect='equal')
plt.title('Prob. of Y = 1')
plt.arrow(0,0,w1[0,0],w1[0,1],width=0.1,length_includes_head=True,color='b',label='w1')
plt.arrow(0,0,w2[0,0],w2[0,1],width=0.1,length_includes_head=True,color='dodgerblue',label='w2')
plt.colorbar()
plt.xlabel('x1')
plt.ylabel('x2')
plt.contour(hw1.T,levels = [0.5],linewidths=1,extent = [-2,2,-2,2])
plt.legend()
plt.tight_layout()
plt.show()
```
